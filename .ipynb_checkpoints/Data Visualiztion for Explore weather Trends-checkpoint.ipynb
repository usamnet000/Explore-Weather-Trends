```python
# import Libraries
import numpy as np
import pandas as pd # For processing large amounts of data 
from matplotlib import pyplot as plt   # plot library I use line chart as the simple

# read DataSet in python
data = pd.read_csv("results.csv")

#Drawing the plot Global Temperature
plt.plot(data['year'], data['GlobalAT'].rolling(142).mean(), label = 'Global')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (°C)")
plt.title("Global Average Temperature")
plt.show()
```

```python
# import Libraries
import numpy as np
import pandas as pd # For processing large amounts of data 
from matplotlib import pyplot as plt   # plot library I use line chart as the simple

# read DataSet in python
data = pd.read_csv("results.csv")

#Drawing the plot Global Temperature
plt.plot(data['year'], data['GlobalAT'].rolling(142).mean(), label = 'Global')
plt.plot(data['year'], data['CityAT'].rolling(142).mean(), label = 'Riyadh')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Temperature (°C)")
plt.title("Compare the temperature in Riyadh and Global")
plt.show()
```
